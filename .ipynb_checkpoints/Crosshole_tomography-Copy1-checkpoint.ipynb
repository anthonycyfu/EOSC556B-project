## Defining a Tensor Mesh

Here, we create the tensor mesh that will be used to predict arrival time 
data. The horizontal and vertical extents are 30m and 50m respectively.  Cell size is 0.5m




```python
# Setting up a mesh

extent_x = 30 
extent_y = 50

dx = .5
dy = .5
```

```python
nx = int(np.ceil(extent_x / dx))
ny = int(np.ceil(extent_y / dy))

hx = np.ones(nx) * dx
hy = np.ones(ny) * dy
```

```python
mesh = discretize.TensorMesh([hx, hy])
mesh
```

## Defining the Survey

Here, we define survey that will be used for the forward simulation. The survey
consists of a vertical line of point sources at x = 0m and a vertical
line of point receivers at x = 30m. The spacing of sources/receivers is 1m. The shot by each source is measured by
all receivers.  




```python
n_src = 50
y_src = np.linspace(dy/2, extent_y-dy/2, n_src)

source_locations = np.c_[np.zeros_like(y_src), y_src]
receiver_locations = np.c_[np.zeros_like(y_src) + mesh.nodes_x[-1], y_src]
```

```python
# Plot the mesh layout

fig, ax = plt.subplots(1, 1, figsize=(8, 10))

mesh.plot_grid(ax=ax, color="k")
ax.set_aspect(1)

ax.plot(source_locations[:, 0], source_locations[:, 1], "C3s", ms=4, label ='Source')
ax.plot(receiver_locations[:, 0], receiver_locations[:, 1], "C0v", ms=4, label = 'Receiver')
ax.legend(bbox_to_anchor=(1, 1))
```

## Model and Mapping on Tensor Mesh

Here, we create the velocity model that will be used to predict the data. Since
the physical parameter for straight ray tomography is slowness, we need to define
a model with slowness values. 




```python
# First we build a base model with a top layer of sediments (v=2000 m/s) and a bottop layer of marble (v=5000m/s.
# The sediments/marble interfaces at BH1 (x=0m) and BH2(x=30m) are 30m and 20m respectively. 

v_sediment = 2000
v_marble = 5000
bh1_interface = 30
bh2_interface = 20



def build_base_model(bh1_interface, bh2_interface, v_sediment, v_marble): 
    inds_marble = (

        (mesh.cell_centers[:, 1] < bh1_interface + (bh2_interface - bh1_interface)/ extent_x * mesh.cell_centers[:, 0])
        
    ) 
    velocity_model = v_sediment*np.ones(mesh.n_cells)
    velocity_model[inds_marble] = v_marble
    
    return velocity_model

velocity_model = build_base_model(bh1_interface, bh2_interface, v_sediment, v_marble) 

ref_velocity_sediment = v_sediment*np.ones(mesh.n_cells)
ref_velocity_marble = v_marble*np.ones(mesh.n_cells)
ref_velocity_layered = velocity_model.copy()

```

```python
# Here we define a function to plot the model
def plot_model_survey(
    mesh, model, source_locations, receiver_locations, ax=None, 
    show_raypaths=True, vmin=None, vmax=None
):
    if ax is None: 
        fig, ax = plt.subplots(1, 1, figsize=(4, 5))
    
    cb = plt.colorbar(
        mesh.plot_image(model, ax=ax, pcolor_opts={"vmin":vmin, "vmax":vmax})[0], 
        ax=ax
    )
    
    cb.set_label("velocity (m/s)")
    
    ax.set_aspect(1)

    ax.plot(source_locations[:, 0], source_locations[:, 1], "ws", ms=3)
    ax.plot(receiver_locations[:, 0], receiver_locations[:, 1], "wv", ms=3)
    
    if show_raypaths is True: 
        for i in range(source_locations.shape[0]):
            for j in range(receiver_locations.shape[0]):
                ax.plot(
                    np.r_[source_locations[i, 0], receiver_locations[j, 0]],
                    np.r_[source_locations[i, 1], receiver_locations[j, 1]],
                    color="k",
                    linewidth=0.2,
                    alpha = 0.2
                )
    return ax

plot_model_survey(
    mesh, velocity_model, source_locations, receiver_locations,
    show_raypaths=False
)
```

```python
# Here we add more structures to the base model.

# Define a function to create elliptical features in the model
def add_ellipse(base_model, center, a, b, value):
    inds_ellipse =  (
                    np.sqrt(
                    (center[0] - mesh.cell_centers[:, 0]) ** 2 / a**2
                    + (center[1] - mesh.cell_centers[:, 1]) ** 2 / b**2)
                    < 1
                )
                
    new_model = base_model.copy()
    new_model[inds_ellipse] = value
    
    return new_model


# Add a cavity in the marble layer

cavity_center = np.array([30,10])
cavity_a = 10
cavity_b = 2
velocity_model = add_ellipse(velocity_model, cavity_center, cavity_a, cavity_b, v_sediment)


# Add a trough in the marble layer

trough_center = np.array([15,30])
trough_a = 10
trough_b = 15
velocity_model = add_ellipse(velocity_model, trough_center, trough_a, trough_b, v_sediment)



# Add three floaters in the sediment layer 

floater_center = np.array([0,40])
floater_a = 6
floater_b = 2
velocity_model = add_ellipse(velocity_model, floater_center, floater_a, floater_b, v_marble)

floater2_center = np.array([20,45])
floater2_a = 1
floater2_b = 1
velocity_model = add_ellipse(velocity_model, floater2_center, floater2_a, floater2_b, v_marble)

floater3_center = np.array([15,35])
floater3_a = 1
floater3_b = 1
velocity_model = add_ellipse(velocity_model, floater3_center, floater3_a, floater3_b, v_marble)

```

```python
# Here we create the true slowness model
true_slowness_model = 1/velocity_model # true model

plot_model_survey(
    mesh, velocity_model, source_locations, receiver_locations,
    show_raypaths=False
)
```

```python
# Here we create 4 reference models which we can possibly use during inversion.

ref_slowness_sediment = 1/ref_velocity_sediment   # reference model with v=2000 m/s
ref_slowness_marble = 1/ref_velocity_marble       # reference model with v=5000 m/s
ref_slowness_layered = 1/ref_velocity_layered     # reference model with 2 layers inferred


borehole1_ind =  (
                (mesh.cell_centers[:,0]== mesh.cell_centers[:,0].min() ) &
                 (mesh.cell_centers[:,1] <= floater_center[1] + floater_b) &
                (mesh.cell_centers[:,1] >= floater_center[1] - floater_b)
                 )

borehole2_ind =  (
                (mesh.cell_centers[:,0]== mesh.cell_centers[:,0].max() ) &
                 (mesh.cell_centers[:,1] <= cavity_center[1] + cavity_b) &
                (mesh.cell_centers[:,1] >= cavity_center[1] - cavity_b)
                 )

ref_velocity_layered_BH = ref_velocity_layered.copy()
ref_velocity_layered_BH[borehole1_ind] = v_marble
ref_velocity_layered_BH[borehole2_ind] = v_sediment

ref_slowness_layered_BH = 1/ref_velocity_layered_BH   # reference model with 2 layers inferred and borehole information
```

## Define the Inverse Problem

The inverse problem is defined by 3 things:

    1) Data Misfit: a measure of how well our recovered model explains the field data
    2) Regularization: constraints placed on the recovered model and a priori information
    3) Optimization: the numerical approach used to solve the inverse problem




```python
# We will use the data with 5% relative error for inversion
relative_error=0.05
np.random.seed(7)

data_for_inversion = (
    clean_data + 
    np.random.randn(len(clean_data)) * relative_error * np.abs(clean_data)
)
```

```python
# Define a data object. Uncertainties are specified.
data_object = data.Data(tomo_survey, dobs=data_for_inversion, relative_error=relative_error)
```

```python
# Define the data misfit. Here the data misfit is the L2 norm of the weighted
# residual between the observed data and the data predicted for a given model.
# Within the data misfit, the residual between predicted and observed data are
# normalized by the data's standard deviation.
dmis = data_misfit.L2DataMisfit(data_object, simulation)
```

```python
# Define the regularization (model objective function). L2 norms are used for the smallness and the derivative terms.
# Different alpha values can be assigned to obtain a smoothest model or a balanced model.
# Different reference model can also be specified.

alpha_s = 1
alpha_x = 1
alpha_y = 1
reference_model = ref_slowness_layered_BH

reg_s = regularization.Smallness(mesh, reference_model=ref_slowness_layered)
reg_x = regularization.SmoothnessFirstOrder(mesh, orientation="x",reference_model=reference_model)
reg_y = regularization.SmoothnessFirstOrder(mesh, orientation="y",reference_model=reference_model)

reg = alpha_s * reg_s + alpha_x * reg_x + alpha_y * reg_y
```

```python
# Define how the optimization problem is solved.

opt = optimization.ProjectedGNCG(lower=0.0, maxIter=50)


# Here we define the inverse problem that is to be solved

invprob = inverse_problem.BaseInvProblem(dmis, reg, opt)
```

## Define Inversion Directives

Here we define any directives that are carried out during the inversion. This
includes the cooling schedule for the beta value, stopping
criteria for the inversion and saving inversion results at each iteration.




```python
chifact=0.5
beta0_ratio=100
coolingFactor=1.2

save = directives.SaveOutputDictEveryIteration()
beta_estimate = directives.BetaEstimate_ByEig(beta0_ratio=beta0_ratio)
beta_schedule = directives.BetaSchedule(
    coolingFactor=coolingFactor, coolingRate=1
)
target = directives.TargetMisfit(chifact=chifact)


directives_list = [
    beta_estimate,
    beta_schedule,
    save,
    target
]
```

## Running the Inversion

To define the inversion object, we need to define the inversion problem and
the set of directives. We can then run the inversion.




```python
inv = inversion.BaseInversion(invprob, directives_list)
```

```python
mopt = inv.run(reference_model)
```

## Plotting the Recovered Model




```python
def plot_tikhonov_curves(beta_values, phi_d, phi_m, phid_star=None, iteration=None, ax=None): 
    if ax is None: 
        fig, ax = plt.subplots(1, 3, figsize=(12,3), dpi=200)
    
    ax[0].plot(beta_values, phi_d)
    ax[1].plot(beta_values, phi_m)
    ax[2].plot(phi_m, phi_d)

    if phid_star is not None: 
        ax[0].plot(beta_values, np.ones_like(beta_values) * phid_star, "--k")
        ax[2].plot(phi_m, np.ones_like(beta_values) * phid_star, "--k")

    ax[0].set_ylabel("$\\phi_d$")
    ax[1].set_ylabel("$\\phi_m$")
    ax[2].set_ylabel("$\\phi_d$")
    ax[2].set_xlabel("$\\phi_m$")
    
    if iteration is not None: 
        ax[0].plot(beta_values[iteration], phi_d[iteration], "C3o")
        ax[1].plot(beta_values[iteration], phi_m[iteration], "C3o")
        ax[2].plot(phi_m[iteration], phi_d[iteration], "C3o")

    for a in ax[:2]:
        a.invert_xaxis()
        a.set_xlabel("$\\beta$")
    plt.tight_layout()
    
    return ax
```

## Adding local weight to the borehole locations in the smallness term

```python
# Find the cell indices of the borehole locations
ind_BH = (
                (mesh.cell_centers[:,0]== mesh.cell_centers[:,0].min() ) |
                 (mesh.cell_centers[:,0]== mesh.cell_centers[:,0].max() )
                )
```

```python
# A factor of 10 is applied to the weights in the smallness term in the borehole locations

reg_s2 = regularization.Smallness(mesh, reference_model=ref_slowness_layered_BH)
reg_s2.set_weights(borehole = ind_BH*10)

reg2 = alpha_s * reg_s2 + alpha_x * reg_x + alpha_y * reg_y
```

```python
# Run the inversion again

opt = optimization.ProjectedGNCG(lower=0.0, maxIter=50)

invprob = inverse_problem.BaseInvProblem(dmis, reg2, opt)

chifact=0.5
beta0_ratio=100
coolingFactor=1.2

save = directives.SaveOutputDictEveryIteration()
beta_estimate = directives.BetaEstimate_ByEig(beta0_ratio=beta0_ratio)
beta_schedule = directives.BetaSchedule(
    coolingFactor=coolingFactor, coolingRate=1
)
target = directives.TargetMisfit(chifact=chifact)


directives_list = [
    beta_estimate,
    beta_schedule,
    save,
    target
]

inv = inversion.BaseInversion(invprob, directives_list)
mopt = inv.run(reference_model)
```

## Non-linear inversion


Detailed explanation of the use of mixed lp norms as well as the IRLS procedures can be found in the publication below.
<br>
Fournier, D. and Oldenburg, D.W., 2019, Inversion using spatially variable mixed lp norms. Geophysical Journal International, 218(1), pp.268-282.


```python
# Define the regularization (model objective function). Here, 'p' defines
# the norm of the smallness term and 'q' defines the norm of the derivative
# term.  They can be any values from 0 to 2.

reg_sparse = regularization.Sparse(mesh, mapping=maps.IdentityMap(nP=mesh.nC))
p = 0
qx = 1
qy = 1
reg_sparse.norms = [p, qx, qy]

opt = optimization.ProjectedGNCG(lower=0.0, maxIter=80)

```

```python
inv_prob_sparse = inverse_problem.BaseInvProblem(dmis, reg_sparse, opt)
```

```python
update_IRLS = directives.Update_IRLS(
    f_min_change=1e-4,
    max_irls_iterations=30,
    coolEpsFact=1.5,
    beta_tol=1e-2,
)

# Defining a starting value for the trade-off parameter (beta) between the data
# misfit and the regularization.
starting_beta = directives.BetaEstimate_ByEig(beta0_ratio=1e5)

# Save output at each iteration
saveDict = directives.SaveOutputDictEveryIteration()

# Define the directives as a list
directives_list = [starting_beta, update_IRLS, saveDict]
```

```python
inv = inversion.BaseInversion(inv_prob_sparse, directives_list)
```

```python
recovered_model = inv.run(ref_slowness_layered_BH)
```
